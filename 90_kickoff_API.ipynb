```python
import pandas as pd
import requests
from datetime import datetime, timedelta
import time

# Liste Ihrer CSV-Dateien
filenames = ["2019-2020.csv", "2020-2021.csv", "2021-2022.csv"]

# Ein Set zum Speichern aller einzigartigen Teams über alle Dateien hinweg
all_unique_teams = set()

# Ein Dictionary zum Speichern der Teams pro Datei (optional, aber von Ihnen angefragt)
teams_by_file = {}

print("Reading date and extrating Team name...")

for filename in filenames:
    try:
        df = pd.read_csv('/content/drive/MyDrive/archive-2/Datasets/' + filename)

        # Find unique teams in the HomeTeam and AwayTeam columns
        home_teams = df['HomeTeam'].unique()
        away_teams = df['AwayTeam'].unique()

        # Combine the teams and adding them to the set.
        file_teams = set(home_teams) | set(away_teams)

        # Saving teams
        teams_by_file[filename] = sorted(list(file_teams))

        # Update set
        all_unique_teams.update(file_teams)

        print(f"  - file '{filename}' success.")

    except FileNotFoundError:
        print(f"  - WARNUNG: file '{filename}' not found.")
    except Exception as e:
        print(f"  - Error reading '{filename}': {e}")

# Sort the final list of all teams
all_unique_teams_sorted = sorted(list(all_unique_teams))

print("\n--- Teams per file ---")
for filename, teams in teams_by_file.items():
    print(f"\n{filename}:")
    print(f"  {', '.join(teams)}")

print("\n--- All unique tems) ---")
print(f"({len(all_unique_teams_sorted)} Teams found)")
print(all_unique_teams_sorted)
```

```python
import pandas as pd
import numpy as np
import requests
import time
from datetime import datetime, timedelta

API_KEY = ""

stadium_coordinates = {
    "Arsenal": (51.555, -0.1086), "Bournemouth": (50.735, -1.838), "Brighton": (50.862, -0.083),
    "Burnley": (53.789, -2.230), "Chelsea": (51.4816, -0.1908), "Crystal Palace": (51.3983, -0.0855),
    "Everton": (53.4389, -2.9664), "Huddersfield": (53.654, -1.768), "Leicester": (52.6203, -1.1422),
    "Liverpool": (53.4308, -2.9608), "Man City": (53.4830, -2.2002), "Man United": (53.4631, -2.2913),
    "Newcastle": (54.9756, -1.6217), "Southampton": (50.9058, -1.3911), "Stoke": (52.9883, -2.1755),
    "Swansea": (51.6422, -3.9347), "Tottenham": (51.6042, -0.0662), "Watford": (51.6497, -0.4014),
    "West Brom": (52.5091, -1.9639), "West Ham": (51.5386, -0.0165), "Cardiff": (51.4728, -3.2031),
    "Fulham": (51.4749, -0.2217), "Wolves": (52.5902, -2.1303), "Aston Villa": (52.5091, -1.8847),
    "Norwich": (52.6222, 1.3092), "Sheffield United": (53.3703, -1.4708), "Leeds": (53.7778, -1.5722),
    "Brentford": (51.4908, -0.2886)
}


def get_owm_weather(lat, lon, timestamp, api_key):
    """Holt historische Wetterdaten für einen exakten UNIX-Zeitstempel."""
    url = f"https://api.openweathermap.org/data/3.0/onecall/timemachine?lat={lat}&lon={lon}&dt={int(timestamp)}&appid={api_key}&units=metric&lang=en"

    try:
        r = requests.get(url)
        if r.status_code != 200:
            # Fallback
            url_25 = f"https://api.openweathermap.org/data/2.5/onecall/timemachine?lat={lat}&lon={lon}&dt={int(timestamp)}&appid={api_key}&units=metric&lang=en"
            r = requests.get(url_25)

        if r.status_code != 200:
            print(f"API Fehler: {r.status_code} - {r.text}")
            return None

        data = r.json()
        if 'data' in data and len(data['data']) > 0:
            current = data['data'][0]

            precip = 0
            if 'rain' in current and isinstance(current['rain'], dict):
                precip += current['rain'].get('1h', 0)
            if 'snow' in current and isinstance(current['snow'], dict):
                precip += current['snow'].get('1h', 0)

            return {
                "temp": current.get('temp'),
                "humidity": current.get('humidity'),
                "wind_speed": current.get('wind_speed'),
                "wind_dir": current.get('wind_deg'),
                "precip": precip,
                "conditions": current['weather'][0]['description'] if 'weather' in current else None
            }
    except Exception as e:
        print(f"Exception beim Request: {e}")
        return None

    return None

# --- Main function ---
def process_season_file(input_file_path, output_file_path):
    print(f"--- Start processing file: {input_file_path} ---")

    try:
        df = pd.read_csv(input_file_path)
    except FileNotFoundError:
        print(f"File not found {input_file_path}")
        return

    # List initialisation
    start_temps, start_winds, start_wind_dirs, start_hums, start_precips, start_conds = [], [], [], [], [], []
    end_temps, end_winds, end_wind_dirs, end_hums, end_precips, end_conds = [], [], [], [], [], []

    all_lists = [start_temps, start_winds, start_wind_dirs, start_hums, start_precips, start_conds,
                 end_temps, end_winds, end_wind_dirs, end_hums, end_precips, end_conds]

    api_calls_count = 0
    max_calls_safe = 950 # Buffer for daily free limits

    for index, row in df.iterrows():
        if api_calls_count >= max_calls_safe:
            print(f"ACHTUNG: Limit reached ({api_calls_count}). Stopp.")
            break

        team = row['HomeTeam']
        date_str = row['Date']
        time_str = row['Time']

        # 1. Check coordinates
        if team not in stadium_coordinates:
            print(f"Coordinates for team {team} not found, skipping...")
            for l in all_lists: l.append(np.nan)
            continue

        lat, lon = stadium_coordinates[team]

        # 2. Calculate time
        try:
            try:
                full_dt_start = datetime.strptime(f"{date_str} {time_str}", '%d/%m/%Y %H:%M')
            except ValueError:
                full_dt_start = datetime.strptime(f"{date_str} {time_str}", '%Y-%m-%d %H:%M')

            full_dt_end = full_dt_start + timedelta(minutes=90)

            # Convert to Unix Timestamp
            ts_start = full_dt_start.timestamp()
            ts_end = full_dt_end.timestamp()

        except Exception as e:
            print(f"Datumsfehler Zeile {index}: {e}")
            for l in all_lists: l.append(np.nan)
            continue

        # 3. API call

        # Call 1: Kickoff
        w_start = get_owm_weather(lat, lon, ts_start, API_KEY)
        api_calls_count += 1

        # Call 2: Fulltime
        w_end = get_owm_weather(lat, lon, ts_end, API_KEY)
        api_calls_count += 1

        if w_start and w_end:
            # Start values
            start_temps.append(w_start['temp'])
            # Calculate m/s to km/h
            start_winds.append(round(w_start['wind_speed'] * 3.6, 2) if w_start['wind_speed'] else 0)
            start_wind_dirs.append(w_start['wind_dir'])
            start_hums.append(w_start['humidity'])
            start_precips.append(w_start['precip'])
            start_conds.append(w_start['conditions'])

            # Ende Werte
            end_temps.append(w_end['temp'])
            end_winds.append(round(w_end['wind_speed'] * 3.6, 2) if w_end['wind_speed'] else 0)
            end_wind_dirs.append(w_end['wind_dir'])
            end_hums.append(w_end['humidity'])
            end_precips.append(w_end['precip'])
            end_conds.append(w_end['conditions'])
        else:
            # Fallback: NaNs
            for l in all_lists:
                if len(l) <= index:
                    l.append(np.nan)
            print(f"Data gap at Index {index}")

        # Show progress
        if index % 20 == 0:
            print(f"Progress: {index} Games. Calls: {api_calls_count}")

        time.sleep(0.15)

    # 5. Saving
    rows_processed = len(start_temps)
    df_result = df.iloc[:rows_processed].copy()

    df_result['Start_Temp_C'] = start_temps
    df_result['Start_Wind_kmh'] = start_winds
    df_result['Start_Wind_Degree'] = start_wind_dirs
    df_result['Start_Humidity'] = start_hums
    df_result['Start_Precip_mm'] = start_precips
    df_result['Start_Conditions'] = start_conds

    df_result['End_Temp_C'] = end_temps
    df_result['End_Wind_kmh'] = end_winds
    df_result['End_Wind_Degree'] = end_wind_dirs
    df_result['End_Humidity'] = end_hums
    df_result['End_Precip_mm'] = end_precips
    df_result['End_Conditions'] = end_conds

    df_result.to_csv(output_file_path, index=False)
    print(f"Fertig! Gespeichert unter: {output_file_path}")
    print(f"Verbrauchte Calls: {api_calls_count}")

process_season_file('/content/drive/MyDrive/archive-2/Datasets/2019-2020.csv', '2019-20_weather_90_full.csv')
```
